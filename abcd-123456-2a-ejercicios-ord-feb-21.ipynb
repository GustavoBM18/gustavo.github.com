Durante la semana de exámenes, los alumnos suelen desvelarse. Se les preguntó cuántas horas
dormían y éstos son los resultados:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5)) 

marcas_clase = [2.6, 3.8, 5, 6.2, 7.4] # Datos del eje X
frecuencias = [2, 3, 7, 10, 0] # Datos actualizados del eje y (he agregado un 0 para que coincida con el número de marcas de clase)

plt.bar(marcas_clase, frecuencias,
        width=1, edgecolor="k", 
        color=["#0F0000", "#FEA029", "#36A77E", "#36A77E", "#4D9B7F"])

plt.xlabel("Horas de sueño", fontsize=25)
plt.ylabel("Numero de alumnos", fontsize=25)
plt.title("Histograma", fontsize=40)
plt.grid()
plt.show()

```

¿Cuántos alumnos contestaron? 25

¿Cuál es la clase más alta y cuál es la más baja? la mas alta es 3.8 horas y la mas baja es 7.4

¿Hacia dónde está la concentración de la distribución? 3.8 horas 

¿Qué significa esto? en que los alumnos no duermen bien en temporada de examenes 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6)) 

marcas_texto = ["Tetra Pack", "Lata", "Botella de plastico", "Botella retornable"] # Etiquetas para las marcas de clase
frecuencias = [1, 2, 5, 4] # Datos del eje y
marcas_clase = range(len(marcas_texto)) # Datos del eje X

# Configuración del diagrama de barras
plt.barh(marcas_clase, frecuencias, height=0.7, edgecolor="k", color=["#0F0000", "#41D4A1", "#38B88C", "#36A77E"])

plt.xlabel("Frecuencia absoluta", fontsize=25) # Etiqueta del eje X
plt.ylabel("Marcas de clase", fontsize=25) # Etiqueta del eje Y
plt.title("Diagrama de barras", fontsize=40) # Título
plt.yticks(marcas_clase, marcas_texto, fontsize=15) # Etiquetas del eje Y
plt.xticks(fontsize=15) # Tamaño de las etiquetas del eje X
plt.grid(axis='x') # Activar cuadrícula solo en el eje X
plt.show() # Mostrar gráfico

```
